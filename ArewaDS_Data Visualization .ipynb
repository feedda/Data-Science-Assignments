## I. Introduction
**Lungs Capacity Dataset:<br>**
Lung capacity refers to the total amount of air that our lungs can hold **(Delgado BJ, Bajaj T, 2023)**, it is measured in terms of volume and is typically assessed through various pulmonary function tests. Lung capacity is a critical indicator of respiratory health, and understanding the factors that influence it, is essential for promoting overall well-being **(Ph.H Quanjer, et al., 1993)**. This project aims to investigate the relationships between lung capacity and several key variables, including smoking habits, age, gender, mode of birth, and height. The project will explore whether these factors have a significant impact on lung capacity, contributing valuable insights to public health and clinical practices.

# 2. Exploratory Data Analysis (EDA)
This is an essential phase in the data analysis process where the main goal is to understand the main characteristics of a dataset. According to **Tukey, (1977)**, EDA involves summarizing the main features, often with the help of graphical representations, to gain insights and identify patterns or trends. EDA can be conducted at various levels, including:**<br>
- Univariate
- Bivariate, and
- Multivariate analysis<br>

**Let's start our EDA with Univariate**

# 2.1 EDA using Univariate Analysis
Univariate analysis focuses on examining the distribution and properties of a single variable in isolation **(Canova, 2017)**, it uses the following statistical analysis:
- Descriptive statistics
- Histograms and Distributions

#### Below are some of the graphical representation of the variables based on the univeriate analysis

```python
# Creating a 2x3 grid of subplots with a specified figure size
fig, axes = plt.subplots(2, 3, figsize=(16, 10))

# Plotting a countplot for the 'Caesarean' variable in the first subplot
sns.countplot('Caesarean', data=data, ax=axes[0, 0])

# Plotting a countplot for the 'Smoke' variable in the second subplot
sns.countplot('Smoke', data=data, ax=axes[0, 1])

# Plotting a countplot for the 'Gender' variable in the third subplot
sns.countplot('Gender', data=data, ax=axes[0, 2])

# Plotting a distribution plot for the 'Height(inches)' variable in the fourth subplot
sns.distplot(data['Height(inches)'], ax=axes[1, 0])

# Plotting a distribution plot for the 'LungCap(cc)' variable in the fifth subplot
sns.distplot(data['LungCap(cc)'], kde=False, ax=axes[1, 1])

# Plotting a distribution plot for the 'Age( years)' variable in the sixth subplot
sns.distplot(data['Age( years)'], kde=True, ax=axes[1, 2])

# Displaying the subplots
plt.show()
```

**The above code is using Matplotlib and Seaborn libraries to create a grid of subplots and visualize different aspects of a dataset, making it easier to compare and analyze multiple variables at once.**<br>

**Now, the following code will summarize the distribution of categorical Variables using pie charts.**

```python
# Creating a 1x3 grid of subplots with a specified figure size
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# Plotting a pie chart for the 'Gender' variable in the first subplot
data['Gender'].value_counts().plot.pie(explode=[0, 0], autopct='%1.0f%%', ax=axes[0], shadow=True)

# Plotting a pie chart for the 'Smoke' variable in the second subplot
data['Smoke'].value_counts().plot.pie(explode=[0.2, 0], autopct='%1.0f%%', ax=axes[1], shadow=True)

# Plotting a pie chart for the 'Caesarean' variable in the third subplot
data['Caesarean'].value_counts().plot.pie(explode=[0.1, 0], autopct='%1.0f%%', ax=axes[2], shadow=True)

# Displaying the subplots
plt.show()
```

The above code created a single row with three pie charts to visualize the distribution of categorical variables in the dataset. Each pie chart represents the proportion of different categories within a specific variable. The explode, autopct, and shadow parameters add visual enhancements to the charts.

Looking at the Charts, we can say that Men are more than women, Smokers are less when compared to Non-smokers and people who born by Caesarean are less when compared with people not born by Caesarean

# 2.2 EDA using Bivariate Analysis
Bivariate analysis explores the relationship between two variables, examining how changes in one variable are related to changes in another **(Bertani et al., 2018)**, it uses the following statistical analysis:
- Scatter Plots
- Correlation Coefficients
- Categorical Plots, etc.

#### Below are some of the graphical representation of the variables based on the biveriate analysis

### Gender Vs Age

```python
# Creating a figure with a specified figure size
fig = plt.figure(figsize=(15, 6))

# Creating a count plot for the 'Gender' variable with 'Age( years)' as hue
sns.countplot(x='Gender', data=data, hue='Age( years)', palette='Accent')

# Adding a title to the plot
plt.title('GENDER vs AGE')

# Displaying the plot
plt.show()
```

**The provided code above utilizes the Matplotlib and Seaborn libraries to create a count plot, visualizing the distribution of age within different gender categories from the dataset. The figure clearly indicates that males aged 12 and 13 are more prevalent, while those aged 4 are less common. Among females, those aged 13 are more abundant, whereas ages 3 and 4 are less frequent.**

### Gender vs Smoke

```python
# Creating a figure with a specified figure size
fig = plt.figure(figsize=(8, 6))

# Generating a count plot for 'Gender' vs 'Smoke'
sns.countplot(x='Gender', data=data, hue='Smoke')

# Adding a title to the plot
plt.title('Gender vs Smoke')

# Displaying the plot
plt.show()
```

**In the provided code above, a count plot is created to visualize the distribution of smoking habits across different gender categories in the Lung Capacity dataset.**

**Here is the Interpretation of the plot:**

1. **Men and Non-Smoking:**
   - Among men, the count of non-smokers is notably higher compared to smokers.
   - This suggests that a larger proportion of men in the dataset do not smoke.

2. **Women and Non-Smoking:**
   - Similarly, among women, the count of non-smokers surpasses that of smokers.
   - The plot indicates that a significant number of women in the dataset are non-smokers.

3. **Comparison of Non-Smokers:**
   - Overall, the count plot demonstrates that non-smokers, irrespective of gender, are more prevalent than smokers in the given Lung Capacity dataset.

**In summary, the count plot effectively illustrates the distribution of smoking habits across gender categories, emphasizing that non-smokers outnumber smokers in the dataset.**

### Gender vs Caesarean 

```python
# Creating a figure with a specified figure size
fig = plt.figure(figsize=(10, 6))

# Generating a cross-tabulation (crosstab) of 'Gender' and 'Caesarean'
ct = pd.crosstab(index=data['Gender'], columns=data['Caesarean'])

# Plotting bars for individuals born by Caesarean
plt.bar(ct.index, ct['yes'], label='Born by Caesarean')

# Plotting bars for individuals not born by Caesarean, stacked on top of the previous bars
plt.bar(ct.index, ct['no'], bottom=ct['yes'], label='Not Born by Caesarean')

# Adding labels and title to the plot
plt.xlabel('Gender')
plt.title('GENDER vs CAESAREAN', color='purple')

# Adding a legend to distinguish between born and not born by Caesarean
plt.legend()

# Displaying the plot
plt.show()
```

In the provided code above, a bar chart is generated to illustrate the distribution of individuals born by Caesarean and those not born by Caesarean across different gender categories in the dataset.

**Interpretation of the presented chart:**

1. **Comparison of Men and Women Not Born by Caesarean:**
   - The bar chart reveals that men not born by Caesarean outnumber women not born by Caesarean.
   - This suggests a notable gender-based difference in the distribution of individuals not born by Caesarean.

2. **Representation of Caesarean Birth:**
   - The chart provides a clear distinction between individuals born by Caesarean (depicted by the lower bars) and those not born by Caesarean (depicted by the stacked upper bars).

**In summary, the bar chart effectively communicates the distribution of individuals born by Caesarean and those not born by Caesarean across different gender categories. The distinction between the two categories is visually apparent, and the chart allows for a quick comparison of gender-based differences in Caesarean birth.**

### Smoke vs Caesarean

```python
# Creating a figure with a specified figure size
fig = plt.figure(figsize=(10, 6))

# Generating a cross-tabulation (crosstab) of 'Caesarean' and 'Smoke'
ct = pd.crosstab(index=data['Caesarean'], columns=data['Smoke'])

# Plotting bars for smokers
plt.bar(ct.index, ct['yes'], label='Smokers')

# Plotting bars for non-smokers, stacked on top of the previous bars
plt.bar(ct.index, ct['no'], bottom=ct['yes'], label='Non-Smokers')

# Adding labels and title to the plot
plt.xlabel('Caesarean')
plt.title('SMOKE vs CAESAREAN', color='darkblue')

# Adding a legend to distinguish between smokers and non-smokers
plt.legend()

# Displaying the plot
plt.show()
```

In the provided code above, a bar chart is created to visualize the distribution of smokers and non-smokers among individuals born by Caesarean and those not born by Caesarean.

**Here is the interpretation of the bar chart:**

1. **Comparison of Non-Smokers Born by Caesarean and Not Born by Caesarean:**
   - The bar chart highlights that non-smokers not born by Caesarean outnumber non-smokers born by Caesarean.
   - This suggests a notable difference in the distribution of non-smokers based on the mode of delivery.

2. **Representation of Smoking Habits:**
   - The chart provides a clear representation of individuals born by Caesarean (depicted by the lower bars) and those not born by Caesarean (depicted by the stacked upper bars) for both smokers and non-smokers.

**In summary, the bar chart effectively communicates the distribution of smokers and non-smokers across different categories of Caesarean birth. It visually emphasizes the contrast in the number of non-smokers based on the mode of delivery and provides insight into the number of individuals, categorized by smoking habits, born by Caesarean or not.**

# 2.3 EDA using Multivariate Analysis
Multivariate analysis involves the study of more than two variables simultaneously. It is used to understand complex relationships and patterns in data involving multiple variables **(Hair et al., 2010)**, It uses the following statistical analysis:
  - Heatmaps
 
#### Below are some of the graphical representation of the variables based on the multivariate analysis

### Heat Map for the Non-categorical Variables in the Dataset (Height, Age and LungCap)

```python
# Creating a figure with a specified figure size
fig = plt.figure(figsize=(10, 8))

# Generating a heatmap based on the correlation matrix of the dataset
sns.heatmap(data.corr(), cmap='Blues')

# Adding a title to the heatmap
plt.title('HEAT MAP', color='darkblue')

# Displaying the heatmap
plt.show()
```

A heat map is a graphical representation that provides a visual summary of data in a two-dimensional format **(Wilkinson & Friendly, 2009)**. In the given code above, a heat map is generated to visually represent the correlation between numeric variables in the dataset .

**Here is the interpretation of the above Heat Map:**

1. **Correlation Representation:**
   - The heatmap is generated based on the correlation matrix of the dataset.
   - Only numeric variables are considered in the correlation analysis.
   - The intensity of colors represents the degree and direction of correlation between variables.

2. **Visualizing Correlation Strength:**
   - Darker shades of blue indicate stronger positive correlations, while lighter shades represent weaker or negative correlations.
   - The color intensity allows for a quick visual assessment of which variables are positively, negatively, or weakly correlated.

In summary, the heat map serves as a powerful tool to visually explore and interpret the correlation structure within the dataset. It provides an immediate and intuitive representation of the relationships between numeric variables, aiding in the identification of patterns and potential insights.
